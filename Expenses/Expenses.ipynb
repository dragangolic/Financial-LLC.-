```python
import pandas as pd
```

```python
# Creating dataset of expenses Jan-Apr
expenses = pd.DataFrame(
{
    "Rent":[1500, 4000, 7654, 9000],
    "Bussines_trip":[25000, 31005, 98000, 10000],
    "It_equipment":[47000, 12000, 2354, 84562],
    "Office_stuff":[2154, 5624, 584, 4125],
    "Food":[3000, 3000, 3000, 3000]
})
```

```python
expenses
```

```python
# Creating dataset of expenses Mai-Avg
expenses_2 = pd.DataFrame (
{
    "Rent":[2100, 4320, 7054, 3500],
    "Bussines_trip":[2500, 1005, 8000, 1000],
    "It_equipment":[1000, 1200, 2001, 562],
    "Office_stuff":[2104, 2524, 2584, 2125],
    "Food":[1000, 1230, 1100, 900]
})
```

```python
# Creating dataset of expenses sep-dec
expenses_3 = pd.DataFrame(
{
    "Rent":[1590, 2000, 6654, 8700],
    "Bussines_trip":[2000, 3005, 8000, 6000],
    "It_equipment":[600, 2000, 3354, 4562],
    "Office_stuff":[2174, 7624, 574, 4725],
    "Food":[2700, 2700, 3000, 2543]
})
```

```python
# Concatinate all datasets
frames = [expenses, expenses_2, expenses_3]
expenses_total = pd.concat((frames), ignore_index=True)
expenses_total
```

```python
# Get sum of rows and isert new column
expenses_total['Month_Sum'] = expenses_total.sum(axis=1)
```

```python
expenses_total
```

```python
# Total expenses per month
total_expenses = expenses_total.set_axis(['Jan', 'Feb', 'Mar', 'Apr', 'Mai', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
total_expenses
```

```python
# Add total row
total_expenses.loc['Total']= total_expenses.sum()
```
